```python
class Firewall:
    def __init__(self):
        self.rules = {}

        self.default_action = "DENY"
        
    def add_rule(self, value, rule_type, action="DENY"):
        if(self.rules.get(value) != None):
            print("Esa regla ya existe!")

        self.rules.update({value : [rule_type, action] })

    def process_packet(self, src_ip, dst_ip, dst_port):
        print(f"Procesando paquete: {src_ip} -> {dst_ip}:{dst_port}")
        
        if (self.rules.get(src_ip) != None):
            print(f"Regla aplicada: valor={src_ip}, tipo de regla= {self.rules.get(src_ip)[0]}, accion= {self.rules.get(src_ip)[1]}")

            return True if self.rules.get(src_ip)[1] == "ALLOW" else False
        elif (self.rules.get(dst_port) != None):
            print(f"Regla aplicada: valor={dst_port}, tipo de regla= {self.rules.get(dst_port)[0]}, accion= {self.rules.get(dst_port)[1]}")

            return True if self.rules.get(dst_port)[1] == "ALLOW" else False
        
        return False
```

```python

# Configuración y pruebas del firewall
def main():
    firewall = Firewall()
    
    # Definir reglas
    firewall.add_rule('192.168.1.100', 'src_ip', 'ALLOW')  # Permitir IP específica
    firewall.add_rule('22', "dst_port", 'ALLOW')             # Permitir HTTP
    firewall.add_rule('443', "dst_port", 'ALLOW')            # Permitir HTTPS
    firewall.add_rule('10.0.0.5', 'src_ip', 'DENY')        # Bloquear IP específica
    
    result1 = firewall.process_packet('192.168.1.100', '192.168.1.1', "8080")
    print(f"Resultado: {'PERMITIDO' if result1 else 'DENEGADO'}\n")
    
    result2 = firewall.process_packet('192.168.1.50', '192.168.1.1', "443")
    print(f"Resultado: {'PERMITIDO' if result2 else 'DENEGADO'}\n")
    
    result3 = firewall.process_packet('10.0.0.5', '192.168.1.1', "80")
    print(f"Resultado: {'PERMITIDO' if result3 else 'DENEGADO'}\n")
    
    result4 = firewall.process_packet('192.168.1.50', '192.168.1.1', "22")
    print(f"Resultado: {'PERMITIDO' if result4 else 'DENEGADO'}\n")

    result5 = firewall.process_packet("0.0.0.1", "132.145.78.1", "444")
    print(f"Resultado: {"PERMITIDO" if result5 else "DENEGADO"}\n")

if __name__ == "__main__":
    main()
```
